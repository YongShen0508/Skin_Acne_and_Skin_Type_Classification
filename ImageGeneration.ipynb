```python
import tensorflow as tf
import numpy as np
import os
import cv2
import matplotlib.pyplot as plt

# Define dataset path
level2_dir = "dataset/train/Level_2"

# Load images and preprocess
def load_images(directory, img_size=(128, 128)):
    images = []
    for filename in os.listdir(directory):
        img_path = os.path.join(directory, filename)
        img = cv2.imread(img_path)
        img = cv2.resize(img, img_size)  # Resize to 128x128
        img = img / 255.0  # Normalize to [0,1]
        images.append(img)
    return np.array(images)

# Load Level_2 images
level2_images = load_images(level2_dir)
print(f"Loaded {level2_images.shape[0]} Level_2 images.")
```

```python
from tensorflow.keras.layers import Dense, Reshape, Conv2DTranspose, BatchNormalization, LeakyReLU

def build_generator():
    model = tf.keras.Sequential([
        Dense(16 * 16 * 256, activation="relu", input_shape=(100,)),  # Input noise
        Reshape((16, 16, 256)),  # Reshape to 16x16x256
        Conv2DTranspose(128, (3,3), strides=2, padding='same', activation='relu'),
        BatchNormalization(),
        Conv2DTranspose(64, (3,3), strides=2, padding='same', activation='relu'),
        BatchNormalization(),
        Conv2DTranspose(3, (3,3), strides=2, padding='same', activation='tanh')  # Output (128,128,3)
    ])
    return model

generator = build_generator()
generator.summary()
```

```python
from tensorflow.keras.layers import Conv2D, Flatten, Dropout

def build_discriminator():
    model = tf.keras.Sequential([
        Conv2D(64, (3,3), strides=2, padding='same', activation=LeakyReLU(0.2), input_shape=(128, 128, 3)),
        Dropout(0.3),
        Conv2D(128, (3,3), strides=2, padding='same', activation=LeakyReLU(0.2)),
        Dropout(0.3),
        Flatten(),
        Dense(1, activation='sigmoid')  # Binary classification (real or fake)
    ])
    return model

discriminator = build_discriminator()
discriminator.summary()
```

```python
from tensorflow.keras.optimizers import Adam

# Create GAN model
def build_gan(generator, discriminator):
    discriminator.compile(optimizer=Adam(learning_rate=0.0002), loss="binary_crossentropy", metrics=["accuracy"])
    discriminator.trainable = False  # Freeze discriminator when training GAN
    gan = tf.keras.Sequential([generator, discriminator])
    gan.compile(optimizer=Adam(learning_rate=0.0002), loss="binary_crossentropy")
    return gan

gan = build_gan(generator, discriminator)
gan.summary()
```

```python
import random

def train_gan(epochs=5000, batch_size=32):
    for epoch in range(epochs):
        # Select random real images
        idx = np.random.randint(0, level2_images.shape[0], batch_size)
        real_images = level2_images[idx]
        
        # Generate fake images
        noise = np.random.normal(0, 1, (batch_size, 100))
        fake_images = generator.predict(noise)

        # Train Discriminator
        real_labels = np.ones((batch_size, 1))
        fake_labels = np.zeros((batch_size, 1))

        d_loss_real = discriminator.train_on_batch(real_images, real_labels)
        d_loss_fake = discriminator.train_on_batch(fake_images, fake_labels)
        d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)

        # Train Generator (via GAN)
        noise = np.random.normal(0, 1, (batch_size, 100))
        g_loss = gan.train_on_batch(noise, real_labels)

        # Print progress
        if epoch % 500 == 0:
            print(f"Epoch {epoch} - D Loss: {d_loss[0]:.4f}, G Loss: {g_loss:.4f}")

train_gan()
```

```python
noise = np.random.normal(0, 1, (5, 100))  # Generate 5 random images
generated_images = generator.predict(noise)

# Plot the generated images
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 5, figsize=(15,3))
for i, img in enumerate(generated_images):
    axes[i].imshow((img + 1) / 2)  # Convert from [-1,1] to [0,1]
    axes[i].axis("off")
plt.show()
```
